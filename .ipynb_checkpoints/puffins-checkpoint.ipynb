```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
#Import the data:
#Make sure the csv file "puffins.csv" is in the same directory.
data = pd.read_csv("puffins.csv")

#View the data:
data
```

```python
#Deplete the data that does not include the complete cloumn information:
cleaned_data = data.dropna(axis=0, how='any')
```

```python
#Replace the string data in the factor "species" with integer:
#Set up the dictionary for replacing the data in "species" column:
replacement_map = {"Arctica": 0, "Cirrhata": 1, "Corniculata": 2}

#Replace the string information with integers:
cleaned_data.loc[:, "species"] = cleaned_data["species"].replace(replacement_map)

cleaned_data["species"] = cleaned_data["species"].astype(int)

#View the data after the replacement:
cleaned_data["species"]
```

```python
#Check the corelation between each factors for setting up simulation model:
#Because the number of "species" only represent the categories of the species, hence we deplete the column from the data set:
factor_data = cleaned_data.drop(columns = "species")

#Compare the corelation between each factors in "factor_data":
corr = factor_data.corr()

#Present the corelation between each factors is displayed by heatmap:
sns.heatmap(corr, vmin=-1.0, vmax=1.0, cmap="RdBu")


#According to the corelation between each factors, only "beak_depth_mm" possess negative rorelation with other factors.
```

```python
#To investigate which factors are capable to distinguish each species, scatter plot were used to present the distribution of 
#each data:
sns.pairplot(cleaned_data, hue = 'species',palette="bright")
#According to the result, the corelation of beak depth(mm) could not specify the data set by species.
#We will further utilise machine learning to generate simulation model for predicting data distribution, 
#which allow the model to specify the species of each plot.
#Based on the results of heat map and scatter plot, it is indicated that every factors except 
#"beak length(mm)" failed to seperate different species. 
#Hence, I will further generate training and testing dataset for simulation models with all factors or 
#without one of the factor to testfy this hypothesis.
```

```python
#In order to generate a simulation model that all the factors in the dataset are considered, analysation pipeline that includes
#standard scaler, principal components analysis and K-Neighbors Classifier.
#To perform principal components analysis(PCA), standardise the data set is required:
scaler = StandardScaler()
scaler.fit(factor_data)
```

```python
#Transform the data set into standardise scale:
factor_scaled_raw_data = scaler.transform(factor_data)

#Transform the standardised data set into the form of dataframe:
factor_scaled_data = pd.DataFrame(factor_scaled_raw_data, columns = factor_data.columns)

#Merge the data set with the column of "species" in "cleaned_data":
factor_scaled_data = pd.merge(factor_scaled_data, cleaned_data["species"], left_index = True, right_index = True)

#View the final data set:
factor_scaled_data
```

```python
#Define the data sets that cite in the simulation model:

x1 = factor_scaled_data[["body_mass_kg", "wing_length_m", "beak_length_mm", "beak_depth_mm"]]
x2 = factor_scaled_data[["body_mass_kg", "wing_length_m", "beak_length_mm"]]
x3 = factor_scaled_data[["body_mass_kg", "wing_length_m", "beak_depth_mm"]]
x4 = factor_scaled_data[["body_mass_kg", "beak_length_mm", "beak_depth_mm"]]
x5 = factor_scaled_data[["wing_length_m", "beak_length_mm", "beak_depth_mm"]]

y = factor_scaled_data["species"]
```

```python
#Construct the analysis pipeline for simulation model:
scaled_pca_knn_model = make_pipeline(
    StandardScaler(),
    PCA(n_components=2),  # Dimensionality reduction into 2 components.
    KNeighborsClassifier() 
)

scaled_pca_knn_model
```

```python
#Based on the data set of "factor_scaled_data", we generate testing and training data sets for simulating model:

train_x1, test_x1, train_y1, test_y1 = train_test_split(x1, y, random_state=42)

train_x2, test_x2, train_y2, test_y2 = train_test_split(x2, y, random_state=42)

train_x3, test_x3, train_y3, test_y3 = train_test_split(x3, y, random_state=42)

train_x4, test_x4, train_y4, test_y4 = train_test_split(x4, y, random_state=42)

train_x5, test_x5, train_y5, test_y5 = train_test_split(x5, y, random_state=42)
```

```python
#Fit the model with the training data with all the factor:
scaled_pca_knn_model.fit(train_x1, train_y1)

#Evaluate the model score of simulating model when it tested with testing data(test_x1, test_y1):
score_1 = scaled_pca_knn_model.score(test_x1, test_y1)

#Fit the model with the training data without the factor "beak_depth_mm":
scaled_pca_knn_model.fit(train_x2, train_y2)

#Evaluate the model score of simulating model when it tested with testing data(test_x2, test_y2):
score_2 = scaled_pca_knn_model.score(test_x2, test_y2)

#Fit the model with the training data without the factor "beak_length_mm":
scaled_pca_knn_model.fit(train_x3, train_y3)

#Evaluate the model score of simulating model when it tested with testing data(test_x3, test_y3):
score_3 = scaled_pca_knn_model.score(test_x3, test_y3)

#Fit the model with the training data without the factor "wing_length_mm":
scaled_pca_knn_model.fit(train_x4, train_y4)

#Evaluate the model score of simulating model when it tested with testing data(test_x4, test_y4):
score_4 = scaled_pca_knn_model.score(test_x4, test_y4)

#Fit the model with the training data without the factor "body_mass_kg":
scaled_pca_knn_model.fit(train_x5, train_y5)

#Evaluate the model score of simulating model when it tested with testing data(test_x5, test_y5):
score_5 = scaled_pca_knn_model.score(test_x5, test_y5)


print(score_1, score_2, score_3, score_4, score_5)

#Based on model score, it demonstrate that the model trained by the data set without the factor 
#"beak_depth_mm" or "body_mass_kg" posses higher model score compare with other models.
#Hence, I first investigate the "scaled_pca_knn_model" constructed with (train_x5, train_y5).
```

```python
#To visualise the result will be presented with decision boundary display and scatter plot, but both parts of the results
#require different analysation methods, hence spliting analysation pipeline needs to be seperated into two steps:

#"transformer_steps" includes "StandardScaler" and "PCA":
transformer_steps = scaled_pca_knn_model[:-1]

#"knn_step" includes "KNeighborsClassifier":
knn_step = scaled_pca_knn_model[-1]
```

```python
#Transformed data with "StandardScaler" and "PCA":
transformed_data = pd.DataFrame(transformer_steps.transform(x5))

#Apply "DecisionBoundaryDisplay" with "transformed_data":
DecisionBoundaryDisplay.from_estimator(knn_step, transformed_data, cmap = "inferno")

#Plot all the dot in "transformed_data":
sns.scatterplot(data = transformed_data, x = 0, y = 1, hue = y, palette = "Dark2")

#Although according to the model score and the figure of decision boundary display, it indicate that the current model could 
#predict the species by boundary, but altering the number of neighbors applied in "KNeighborsClassifier" could elevate the 
#model score.
```

```python
#Set up a list of parameters allowed the GridSearchCV to cite in the model:
hyperparameters = {"kneighborsclassifier__n_neighbors" : range(1, 175)}

#Create a list that document all the information that multiple models applied different number of neighbors in "KNeighborsClassifier":
mul_model = GridSearchCV(scaled_pca_knn_model, hyperparameters) 

#Fit the model with the training data without the factor "body_mass_kg":
mul_model.fit(train_x5, train_y5)

best_model = mul_model.best_estimator_
```

```python
#Visualise how the model score varies with the change of number of neighbors in the "KNeighborsClassifier":
cv_results = pd.DataFrame(mul_model.cv_results_) #Define the parameter testing results for each neighbor parameter in the model.
cv_results.rename(columns={"param_kneighborsclassifier__n_neighbors": "n_neighbors"}, inplace=True)
cv_results.plot.scatter("n_neighbors", #Deifne X axis as every neighbour parameter that tested in the model.
                        "mean_test_score", #Define Y axis as test score for corresponded neighbour parameter in the model.
                        yerr="std_test_score", #Define error bar to demonstrate the result of test score result for each parameter.
                        figsize=(10,8) #define the size of the figure.
                       )
```

```python
#Check the number of neighbors with highest model score:
best_model_params = mul_model.best_params_
best_neighbors = best_model_params['kneighborsclassifier__n_neighbors']
print("Best number of neighbors:", best_neighbors, ", Best model score:", best_model.score(test_x5, test_y5))
```

```python
#Create a analysation pipeline with "9" as the number of neighbors in "KNeighborsClassifier":
final_scaled_pca_knn_model = make_pipeline(
    StandardScaler(),
    PCA(n_components=2),  # PCA with 2 components
    KNeighborsClassifier(n_neighbors = 9)
)

final_scaled_pca_knn_model
```

```python
#Fit the analysation pipeline with the training data:
final_scaled_pca_knn_model.fit(train_x5, train_y5)

#Evaluate the model score of simulating model when it tested with testing data.
final_scaled_pca_knn_model.score(test_x5, test_y5)
```

```python
#"transformer_steps" includes "StandardScaler" and "PCA":
final_transformer_steps = final_scaled_pca_knn_model[:-1]

#"knn_step" includes "KNeighborsClassifier":
final_knn_step = final_scaled_pca_knn_model[-1]
```

```python
#Transformed data with "StandardScaler" and "PCA":
final_transformed_data = pd.DataFrame(final_transformer_steps.transform(x5))

#Apply "DecisionBoundaryDisplay" with "transformed_data":
DecisionBoundaryDisplay.from_estimator(final_knn_step, final_transformed_data, cmap = "inferno")

#Plot all the dot of both datasets:
sns.scatterplot(data = final_transformed_data, x = 0, y = 1, hue = y, palette = "Dark2")

#Based on our results, it is demonstrated that although the model score reach 0.9642857142857143, and 
#decision boundary display could roughly specify each species, but the species 1("Cirrhata") spill into other
#boundary of other two species, which is why I further test for another data set with higher model score: "x2". 
```

```python
scaled_pca_knn_model.fit(train_x2, train_y2)

#Set up a list of parameters allowed the GridSearchCV to cite in the model:
hyperparameters = {"kneighborsclassifier__n_neighbors" : range(1, 175)}

#Create a list that document all the information that multiple models applied different number of neighbors in "KNeighborsClassifier":
mul_model = GridSearchCV(scaled_pca_knn_model, hyperparameters) 

#Fit the model with the training data (train_x2, train_y2):
mul_model.fit(train_x2, train_y2)

best_model = mul_model.best_estimator_
```

```python
#Visualise how the model score varies with the change of number of neighbors in the "KNeighborsClassifier":
cv_results = pd.DataFrame(mul_model.cv_results_) #Define the parameter testing results for each neighbor parameter in the model.
cv_results.rename(columns={"param_kneighborsclassifier__n_neighbors": "n_neighbors"}, inplace=True)
cv_results.plot.scatter("n_neighbors", #Deifne X axis as every neighbour parameter that tested in the model.
                        "mean_test_score", #Define Y axis as test score for corresponded neighbour parameter in the model.
                        yerr="std_test_score", #Define error bar to demonstrate the result of test score result for each parameter.
                        figsize=(10,8) #define the size of the figure.
                       )
```

```python
#Check the number of neighbors with highest model score:
best_model_params = mul_model.best_params_
best_neighbors = best_model_params['kneighborsclassifier__n_neighbors']
print("Best number of neighbors:", best_neighbors, ", Best model score:", best_model.score(test_x2, test_y2))
```

```python
#Create a analysation pipeline with "25" as the number of neighbors in "KNeighborsClassifier":
final_scaled_pca_knn_model_2 = make_pipeline(
    StandardScaler(),
    PCA(n_components=2),  # PCA with 2 components
    KNeighborsClassifier(n_neighbors = 25)
)

final_scaled_pca_knn_model_2
```

```python
#Fit the analysation pipeline with the training data:
final_scaled_pca_knn_model_2.fit(train_x2, train_y2)

#Evaluate the model score of simulating model when it tested with testing data.
final_scaled_pca_knn_model_2.score(test_x2, test_y2)
```

```python
#"transformer_steps" includes "StandardScaler" and "PCA":
final_transformer_steps_2 = final_scaled_pca_knn_model_2[:-1]

#"knn_step" includes "KNeighborsClassifier":
final_knn_step_2 = final_scaled_pca_knn_model_2[-1]
```

```python
#Transformed data with "StandardScaler" and "PCA":
final_transformed_data_2 = pd.DataFrame(final_transformer_steps_2.transform(x2))

#Apply "DecisionBoundaryDisplay" with "transformed_data":
DecisionBoundaryDisplay.from_estimator(final_knn_step_2, final_transformed_data_2, cmap = "inferno")

#Plot all the dot of both datasets:
sns.scatterplot(data = final_transformed_data_2, x = 0, y = 1, hue = y, palette = "Dark2")

#Based on the result, with the numbers of neighbors = 25, the model score maintains as 0.9642857142857143,
#which is equal to the model "final_scaled_pca_knn_model(numbers of neighbors = 9)". 
#Despite some of the individuals in the species "Cirrhata" appears in other species' decision boundary, but 
#with number of neighbors = 25, it will be more restricted when "final_scaled_pca_knn_model_2" decides the 
#boundary boarder.
```

#Data analysis pipeline assessment:
In my analysis pipeline, I view the distribution of each species between scales of different factors' combinations to evaluate which factor fits for classifying different species of puffins.

Because there are more than three factors in the data set, my analysis pipeline includes Princiapal Component Analysis(PCA) to perform dimensionality reduction and KNeighborsClassifer to draw decision boarder. 

To choose which factors combination would fit the most for generating model that specify the puffin species, I create multiple training and testing data sets which contain at least three factors, then test thier model socre to identify which data set is the perfect match for the model. According to the result, the data set "x5"(Includes the factors: "wing_length_m", "beak_length_mm", "beak_depth_mm") and "x2"(Includes the factors: "body_mass_kg", "wing_length_m", "beak_length_mm") got higher model that above 0.95. In these two data set, "x5" possess higher model score, so I further conduct decision boundary display with scatter plot to demonstrate the effectiveness of species classification.

Although according to the model "final_scaled_pca_knn_model", the model score and the figure of decision boundary display indicate that the current model could predict the species by boundary, but altering the number of neighbors applied in "KNeighborsClassifier" could elevate the model score. After utilizing "GridSearchCV" to observe the change of mean model score with the variation of "number of neighbors", "0.9642857142857143"(n_neighbors = 9) is already the best score in this model.

To seek for a model with higher score, I further set up a model with "x2", and attempt to elevate the score by altering the "number of neighbors". In "final_scaled_pca_knn_model_2", the final model score is "0.9642857142857143", which is identical to "final_scaled_pca_knn_model". But with higher number of neighbors, it will be more restricted when "final_scaled_pca_knn_model_2" decides the boundary boarder, which indicate it might be a better model for species classification.
